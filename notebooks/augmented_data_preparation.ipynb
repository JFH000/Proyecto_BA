```python
import pandas as pd
import numpy as np

df = pd.read_csv("Food_Delivery_Times.csv")

# duplicar dataset
df_aug = pd.concat([df]*535, ignore_index=True)

# agregar ruido a variables numéricas
for col in ['Distance_km', 'Preparation_Time_min', 'Delivery_Time_min']:
    df_aug[col] = df_aug[col] * np.random.uniform(0.9, 1.1, size=len(df_aug))

df_aug.to_csv("Augmented_Food_Delivery_Times.csv", index=False)
```
